```python
# Y = a + bX

def leastsquare(dataset):
    x = [point[0] for point in dataset]
    y = [point[1] for point in dataset]

    averagex = np.mean(x) # \overline{x} | average value of the values in the array
    averagey = np.mean(y) # \overline{y}

    # calculate b
    bdenominator = 0 # sum of (x - \overline{x}) * (y - \overline{y})
    bcounter = 0 # sum of (x - \overline{x})^2
    for point in dataset:
        xvalue = point[0]
        yvalue = point[1]
    
        xminusaverage = xvalue - averagex
        yminusaverage = yvalue - averagey

        bdenominator += xminusaverage * yminusaverage
        bcounter += xminusaverage ** 2
    
    b = bdenominator / bcounter
    
    # calculate a
    a = averagey - b * averagex # \overline{y} = a + b * \overline{x}

    # solve formula with all x's and create new x, y pairs
    resulty = []

    for point in dataset:
        xvalue = point[0]
        resulty.append(a + b * xvalue)
    
    # display
    plt.plot(x, resulty)
    plt.show()    
```

```python
leastsquare(data_array_1)
```

```python
leastsquare(data_array_2)
```

```python
leastsquare(data_array_3)
```
